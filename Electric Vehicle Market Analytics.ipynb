● DATA CLEANING & PRE-PROCESSING

</h1>

## **Step 1: Check Data types**

```python
df.dtypes
```

## **Step 2: Handling Missing Values**

```python
df.isnull().sum()
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
df.fillna("Unknown", inplace=True)
df.isnull().sum()
```

## **Step 3: Remove Duplicates**

```python
df.drop_duplicates(inplace=True)

print("Duplicates Removed")
```

## **Step 4: Data formatting**
This step ensures consistency in the dataset by correcting data formats, standardizing text values, and preparing columns for analysis.

```python
df.columns = df.columns.str.strip()

df["Make"] = df["Make"].str.upper()

print("Formatting Completed")
```

## **Step 5: Derived Feautures (Feauture engineering)**
New features are created from existing variables to improve analysis and extract meaningful insights.

```python
df["Vehicle_Age"] = 2026 - df["Model Year"]

print(df[["Model Year","Vehicle_Age"]].head())
```

## **Step 6: Data Transformation**
Transformation converts raw data into a suitable format for analysis and improves data consistency.

```python
df["Electric Range"] = pd.to_numeric(
    df["Electric Range"],
    errors="coerce"
)

print("Transformation Completed")
```

```python
df["Electric Range"].head()
```

```python
df.info()
```

## **Step 7:Outlier Detection**

### **Select Numerical Columns**

```python
df.select_dtypes(include=["int64","float64"]).columns
```

### **Detect Outliers Using Box Plot**

```python
plt.figure(figsize=(8,5))

sns.boxplot(
x=df["Electric Range"]
)

plt.title("Outlier Detection – Electric Range")

plt.show()
```

### **Calculate Outliers using IQR Method**

```python
Q1 = df["Electric Range"].quantile(0.25)

Q3 = df["Electric Range"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR

upper = Q3 + 1.5 * IQR

print("Lower Limit:", lower)

print("Upper Limit:", upper)
```

## **Handle Outliers**

```python
df["Electric Range"] = df["Electric Range"].clip(
lower=lower,
upper=upper
)

print("Outliers handled successfully")
df["Electric Range"].describe()
```

# **UNIVARIATE ANALYSIS**

Univariate analysis is performed to analyze individual variables independently and understand their distribution, characteristics, and overall patterns in the Electric Vehicle dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Q1: Which EV type is most common?**

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x="Electric Vehicle Type")

plt.title("Distribution of Electric Vehicle Types")

plt.xticks(rotation=20)

plt.show()
```

## *Insight*

The count plot shows that Battery Electric Vehicles (BEVs) are significantly more common than Plug-in Hybrid Electric Vehicles (PHEVs) in the dataset. The number of BEVs is substantially higher, indicating stronger adoption and preference for fully electric vehicles compared to hybrid alternatives.

This suggests that consumers are increasingly moving toward fully electric transportation due to environmental awareness, technological improvements, and growing charging infrastructure.

## **Q2: What is the Distribution of Electric Range?**

```python
plt.figure(figsize=(10,5))

sns.histplot(
df["Electric Range"],
bins=30
)

plt.title("Electric Range Distribution")

plt.show()
```

## *Insight*

The histogram shows that most electric vehicles in the dataset have a relatively lower electric driving range, while only a smaller number of vehicles achieve higher electric ranges. The distribution is positively skewed, indicating that high-range electric vehicles are less common.
The concentration of vehicles at lower ranges suggests that many EVs are designed for regular urban and short-distance usage, while fewer models are built for extended travel.

## **Q3: Which manufactures dominate EV Market?**

```python
top_make = df["Make"].value_counts().head(10)

plt.figure(figsize=(8,8))

plt.pie(
top_make,
labels=top_make.index,
autopct="%1.1f%%",
startangle=90
)

plt.title(
"Top 10 Manufacturers in EV Market"
)

plt.show()
```

## *Insight*

The pie chart shows that Tesla holds the largest share among the top electric vehicle manufacturers in the dataset, contributing approximately half of the total distribution among the top 10 manufacturers. Other manufacturers such as Chevrolet, Nissan, Ford, Kia, and Toyota contribute smaller proportions.
This indicates that the EV market is highly concentrated, with a few major manufacturers accounting for a significant portion of electric vehicle adoption.

## **Q4: How are Vehicle model years distributed?**

```python
year_counts = df["Model Year"].value_counts().sort_index()

plt.figure(figsize=(10,5))

plt.plot(
year_counts.index,
year_counts.values,
marker="o"
)

plt.title(
"Distribution of Vehicle Model Years"
)

plt.xlabel(
"Model Year"
)

plt.ylabel(
"Number of Vehicles"
)

plt.show()
```

## *Insight*

The line chart shows a clear upward trend in the number of electric vehicles across model years. Earlier model years contain comparatively fewer vehicles, while recent years show a rapid increase in vehicle counts. The highest concentration appears around the latest model years, indicating growing adoption of electric vehicles.
This pattern suggests that consumer interest and production of electric vehicles have accelerated in recent years due to technological advancements, environmental awareness, and market expansion.

# **BIVARIATE ANALYSIS**
Bivariate analysis is performed to study the relationship between two variables and identify patterns, associations, and trends in the Electric Vehicle dataset.

## **Q5: Does electric range vary by EV type?**

```python
plt.figure(figsize=(10,6))

sns.boxplot(
data=df,
x="Electric Vehicle Type",
y="Electric Range"
)

plt.xticks(rotation=20)

plt.show()
```

## *Insight*

The box plot indicates that electric range differs between Battery Electric Vehicles (BEV) and Plug-in Hybrid Electric Vehicles (PHEV). BEVs show a much wider spread and higher electric range values compared to PHEVs, indicating stronger battery performance and longer driving capability.
PHEVs have a relatively smaller range distribution and lower median range because they rely on both electric power and conventional fuel systems. Several outliers are visible for both categories, representing models with exceptionally high electric ranges.

## **Q6: Which manufacturers provide higher range?**

```python
plt.figure(figsize=(12,6))

sns.barplot(
data=df,
x="Make",
y="Electric Range"
)

plt.xticks(rotation=90)

plt.show()
```

## *Insight*

The bar chart compares the average electric range provided by different vehicle manufacturers. It shows that only a few manufacturers offer significantly higher electric ranges, while most manufacturers provide moderate or lower range values.
Manufacturers such as Volkswagen and some premium EV brands appear to provide higher electric ranges compared with others. Several manufacturers remain concentrated in lower range categories, indicating differences in battery technology, vehicle design, and market positioning.
The variation across manufacturers suggests that electric range is an important factor differentiating EV brands in the market.

## **Q7: Relationship between model year and range?**

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
data=df,
x="Model Year",
y="Electric Range"
)

plt.show()
```

## *Insight*

The scatter plot shows the relationship between vehicle model year and electric range. The distribution indicates that newer vehicle models generally tend to provide higher electric ranges compared to older models.
From around 2018 onwards, there is a noticeable increase in vehicles offering ranges above 200 miles, suggesting improvements in battery efficiency and EV technology. However, some recent models still appear with lower ranges, showing that range performance varies depending on vehicle category and manufacturer.
The spread of points also indicates that electric range is not determined only by model year but may also depend on factors such as manufacturer and vehicle type.

## **Q8: How does Electric Vehicle adoption vary across different Cities?**

```python
plt.figure(figsize=(12,6))

top_cities = df["City"].value_counts().head(10)

sns.barplot(
x=top_cities.index,
y=top_cities.values
)

plt.title("Top 10 Cities by EV Adoption")

plt.xlabel("City")

plt.ylabel("Number of Electric Vehicles")

plt.xticks(rotation=45)

plt.show()
```

## *Insight*

The bar chart shows that EV adoption is not evenly distributed across cities. Seattle recorded the highest number of electric vehicle registrations, significantly exceeding all other cities in the dataset. Cities such as Bellevue, Vancouver, Redmond, and Bothell also show comparatively high EV adoption, while the remaining cities have lower counts.
This suggests that EV adoption is concentrated in major urban areas where charging infrastructure, population density, and accessibility to electric vehicles are stronger.

# **MULTIVARIATE ANALYSIS**
Multivariate analysis is performed to study the relationship among multiple variables simultaneously and identify complex patterns and interactions in the Electric Vehicle dataset.

## **Q9: How does Electric Range vary across Vehicle Types and Model Years?**

```python
plt.figure(figsize=(12,6))

sns.violinplot(
    data=df,
    x="Model Year",
    y="Electric Range"
)

plt.title("Electric Range Distribution Across Model Years")
plt.xlabel("Model Year")
plt.ylabel("Electric Range")

plt.xticks(rotation=45)

plt.show()
```

## *Insight*

The violin plot shows that electric range varies significantly across different model years and vehicle types. Earlier model years (1999–2010) show lower and less consistent electric ranges, whereas vehicles from recent years, especially around 2016–2020, demonstrate higher electric ranges and wider distributions. The highest concentration of electric ranges appears in newer model years, indicating improvements in battery technology and vehicle efficiency over time. However, after 2021, some vehicle models show lower ranges, suggesting the presence of different EV categories with varying battery capacities.

## **Q10: What relationships exist among numerical variables?**

```python
plt.figure(figsize=(10,7))

sns.heatmap(
df.select_dtypes(
include="number"
).corr(),
annot=True,
cmap="coolwarm"
)

plt.title(
"Correlation Matrix"
)

plt.show()
```

## *Insight*

The correlation heatmap shows the relationship between numerical variables in the Electric Vehicle dataset. Among the variables, Model Year and Electric Range (-0.55) exhibit a moderate negative correlation, indicating an inverse relationship between these two variables in the dataset. Postal Code and 2020 Census Tract (0.51) show a moderate positive correlation, suggesting some association between location-related attributes. Additionally, Model Year and DOL Vehicle ID (0.35) display a weak positive relationship. Most of the remaining variables have correlation values close to zero, indicating weak or no significant linear relationship. Overall, the analysis suggests that only a few numerical variables show noticeable associations, while most variables are relatively independent.

Electric Range displays noticeable variation across different model years, while Postal Code appears to have little visible influence on both Model Year and Electric Range, as the data points are widely dispersed without a clear pattern. The diagonal histograms show that most vehicles belong to recent model years and lower-to-moderate electric range categories. Overall, the analysis suggests a stronger association between Model Year and Electric Range, whereas Postal Code shows minimal relationship with the other variables.
